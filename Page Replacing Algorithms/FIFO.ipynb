```python
def fifo(ref, frames):
    memory = []
    miss = 0
    hit=0
    for page in ref:
        if page not in memory:
            if len(memory) < frames:
                memory.append(page)
            else:
                memory.pop(0)
                memory.append(page)
            miss += 1
        else:
          hit +=1
        print(page, "->", memory)

    print('\nTotal Miss =',miss)
    print('\n Total Hit =',hit)



pages = list(map(int, input("Enter pages (space separated): ").split()))
frames = int(input("Enter number of frames: "))

fifo(pages, frames)
```
