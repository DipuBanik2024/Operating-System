```python
def lru(ref, frames):
    memory = []
    miss = 0
    hit = 0

    for page in ref:

        if page in memory:
            hit += 1
            memory.remove(page)
            memory.append(page)
        else:
            miss += 1

            if len(memory) < frames:
                memory.append(page)
            else:
                memory.pop(0)
                memory.append(page)

        print(page, "->", memory)

    print("Miss =", miss)
    print("Hit =", hit)


pages = list(map(int, input("Enter pages (space separated): ").split()))
frames = int(input("imput frame"))

lru(pages, frames)
```
