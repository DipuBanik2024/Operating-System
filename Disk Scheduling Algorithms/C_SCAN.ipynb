```python
def cscan(queue, head, direction, disk_size=199):
    total = 0
    current = head

    left = []
    right = []


    for track in queue:
        if track < head:
            left.append(track)
        else:
            right.append(track)

    left.sort()
    right.sort()

    print("Path:", current, end=" ")

    if direction == "inc":

        for track in right:
            movement = abs(current - track)
            print(f"\nMove: |{current}-{track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")


        if current != disk_size:
            movement = abs(current - disk_size)
            print(f"\nMove to max cylinder: |{current}-{disk_size}| = {movement}")
            total += movement
            current = disk_size
            print(disk_size, end=" ")


        print("\nJump from max to 0")
        current = 0
        print(current, end=" ")


        for track in left:
            movement = abs(current - track)
            print(f"\nMove: |{current}-{track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")

    else:

        left.sort(reverse=True)
        for track in left:
            movement = abs(current - track)
            print(f"\nMove: |{current}-{track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")

        if current != 0:
            movement = abs(current - 0)
            print(f"\nMove to 0: |{current}-0| = {movement}")
            total += movement
            current = 0
            print(current, end=" ")


        print("\nJump from 0 to max cylinder")
        current = disk_size
        print(current, end=" ")


        for track in right:
            movement = abs(current - track)
            print(f"\nMove: |{current}-{track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")

    print("\nTotal head movement =", total)


queue = list(map(int, input("Enter queue: ").split()))
head = int(input("Enter head: "))
direction = input("Enter direction (inc/dec): ")

cscan(queue, head, direction)
```
