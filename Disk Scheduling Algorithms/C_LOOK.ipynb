```python
def clook(queue, head, direction):
    total = 0
    current = head

    left = []
    right = []


    for track in queue:
        if track < head:
            left.append(track)
        else:
            right.append(track)

    left.sort()
    right.sort()

    print("Path:", current, end=" ")

    if direction == "inc":

        for track in right:
            movement = abs(current - track)
            print(f"\nMove: |{current}-{track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")


        if left:
            print("\nJump from last right to first left")
            current = left[0]
            print(current, end=" ")


        for track in left[1:]:
            movement = abs(current - track)
            print(f"\nMove: |{current}-{track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")

    else:
        left.sort(reverse=True)

        for track in left:
            movement = abs(current - track)
            print(f"\nMove: |{current}-{track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")


        if right:
            print("\nJump from last left to first right")
            current = right[0]
            print(current, end=" ")


        for track in right[1:]:
            movement = abs(current - track)
            print(f"\nMove: |{current}-{track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")

    print("\nTotal head movement =", total)



queue = list(map(int, input("Enter queue: ").split()))
head = int(input("Enter head: "))
direction = input("Enter direction (inc/dec): ")

clook(queue, head, direction)
```
