```python
def sstf(queue, head):
    total = 0
    current = head

    print("Path:", current, end=" ")

    while len(queue) > 0:
        nearest = queue[0]
        min_dist = abs(current - nearest)

        for track in queue:
            dist = abs(current - track)
            if dist < min_dist:
                min_dist = dist
                nearest = track


        movement = abs(nearest - current)
        print(f"\nMove: |{current} - {nearest}| = {movement}")

        total += abs(movement)
        current = nearest
        queue.remove(nearest)

        print(nearest, end=" ")

    print("\nTotal head movement =", total)


queue = list(map(int, input("Enter queue: ").split()))
head = int(input("Enter head: "))

sstf(queue, head)
```
