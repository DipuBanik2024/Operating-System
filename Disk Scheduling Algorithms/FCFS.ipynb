```python
def fcfs(queue, head):
    total = 0
    current = head

    print("Path:", head, end=" ")

    for track in queue:
        print(track)

        move = abs(track - current)
        print(f"|{track}-{current}| = {move}")

        total += move
        current = track

    print("\nTotal head movement =", total)


queue = list(map(int, input("Enter numbers a: ").split()))
head = int(input("Enter a number: "))

fcfs(queue, head)
```
