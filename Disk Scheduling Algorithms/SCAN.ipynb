```python
def scan(queue, head, direction, disk_size=199):
    total = 0
    current = head

    left = []
    right = []

    for track in queue:
        if track < head:
            left.append(track)
        else:
            right.append(track)

    left.sort(reverse=True)
    right.sort()

    print("Path:", current, end=" ")

    if direction == "dec":   # decreasing order

        for track in left:
            movement = abs(current - track)
            print(f"\nMove: |{current} - {track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")

        if current != 0:
            movement = abs(current - 0)
            print(f"\nMove: |{current} - 0| = {movement}")
            total += movement
            current = 0
            print(0, end=" ")

        for track in right:
            movement = abs(current - track)
            print(f"\nMove: |{current} - {track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")

    else:   # increasing order

        for track in right:
            movement = abs(current - track)
            print(f"\nMove: |{current} - {track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")

        if current != disk_size:
            movement = abs(current - disk_size)
            print(f"\nMove: |{current} - {disk_size}| = {movement}")
            total += movement
            current = disk_size
            print(disk_size, end=" ")

        for track in left:
            movement = abs(current - track)
            print(f"\nMove: |{current} - {track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")

    print("\nTotal head movement =", total)


queue = list(map(int, input("Enter queue: ").split()))
head = int(input("Enter head: "))
direction = input("Enter direction (inc/dec): ")

scan(queue, head, direction)
```
