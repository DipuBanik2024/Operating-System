# Disk Scheduling Algorithms

```python
def fcfs(q,head):
  total = 0
  current=head
  print('path : --->',head,end='')
  for track in q:
    print('\nCurrent Head :',track)
    move=abs(track-current)
    print(f"|{track}-{current}|={move}")
    total+=move
    current=track
  print('total distance :',total)
#q=list(map(int,input("Enter Number a : ").split()))
#head=int(input("Enter Number b : "))
q=[98,183,37,122,14,124,65,67]
h=53
fcfs(q,h)
```

```python
def sstf(q,h):
  total=0
  current=h
  print('path : --->',h,end='')
  while len(q)>0:
    near=q[0]
    min_dist=abs(near-current)
    for i in q:
      dist=abs(i-current)
      if dist<min_dist:
        min_dist=dist
        near=i

    move=abs(near-current)
    print(f"|{near}-{current}|={move}")
    total+=move
    current=near
    q.remove(near)
    print('path : --->',near,end='')
  print('total distance :',total)
q=[98,183,37,122,14,124,65,67]
h=53
sstf(q,h)
```

```python

def cscan(queue, head, direction, disk_size=199):
    total = 0
    current = head

    left = []
    right = []


    for track in queue:
        if track < head:
            left.append(track)
        else:
            right.append(track)

    left.sort()
    right.sort()

    print("Path:", current, end=" ")

    if direction == "inc":

        for track in right:
            movement = abs(current - track)
            print(f"\nMove: |{current}-{track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")


        if current != disk_size:
            movement = abs(current - disk_size)
            print(f"\nMove to max cylinder: |{current}-{disk_size}| = {movement}")
            total += movement
            current = disk_size
            print(disk_size, end=" ")

        total+=disk_size
        print("\nJump from max to 0")
        current = 0
        print(current, end=" ")


        for track in left:
            movement = abs(current - track)
            print(f"\nMove: |{current}-{track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")
    else:

        left.sort(reverse=True)
        for track in left:
            movement = abs(current - track)
            print(f"\nMove: |{current}-{track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")

        if current != 0:
            movement = abs(current - 0)
            print(f"\nMove to 0: |{current}-0| = {movement}")
            total += movement
            current = 0
            print(current, end=" ")

        total+=disk_size
        print("\nJump from 0 to max cylinder")
        current = disk_size
        print(current, end=" ")

        right.sort(reverse=True)
        for track in right:
            movement = abs(current - track)
            print(f"\nMove: |{current}-{track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")
    print("\nTotal head movement =", total)
q=[98,183,37,122,14,124,65,67]
h=53
d='dec'
cscan(q,h,d)
```

```python
def scan(queue, head, direction, disk_size=199):
    total = 0
    current = head

    left = []
    right = []

    for track in queue:
        if track < head:
            left.append(track)
        else:
            right.append(track)

    left.sort(reverse=True)
    right.sort()

    print("Path:", current, end=" ")

    if direction == "dec":   # decreasing order

        for track in left:
            movement = abs(current - track)
            print(f"\nMove: |{current} - {track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")

        if current != 0:
            movement = abs(current - 0)
            print(f"\nMove: |{current} - 0| = {movement}")
            total += movement
            current = 0
            print(0, end=" ")

        for track in right:
            movement = abs(current - track)
            print(f"\nMove: |{current} - {track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")

    else:   # increasing order

        for track in right:
            movement = abs(current - track)
            print(f"\nMove: |{current} - {track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")

        if current != disk_size:
            movement = abs(current - disk_size)
            print(f"\nMove: |{current} - {disk_size}| = {movement}")
            total += movement
            current = disk_size
            print(disk_size, end=" ")

        for track in left:
            movement = abs(current - track)
            print(f"\nMove: |{current} - {track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")

    print("\nTotal head movement =", total)


queue = [98,183,37,122,14,124,65,67]
head = 53
direction ='inc'
scan(queue, head, direction)
```

```python

def clook(queue, head, direction):
    total = 0
    current = head

    left = []
    right = []


    for track in queue:
        if track < head:
            left.append(track)
        else:
            right.append(track)

    left.sort()
    right.sort()

    print("Path:", current, end=" ")

    if direction == "inc":

        for track in right:
            movement = abs(current - track)
            print(f"\nMove: |{current}-{track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")


        if left:
            total+=abs(current-left[0])
            print("\nJump from last right to first left")
            current = left[0]
            print(current, end=" ")


        for track in left[1:]:
            movement = abs(current - track)
            print(f"\nMove: |{current}-{track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")

    else:
        left.sort(reverse=True)
        right.sort(reverse=True)
        for track in left:
            movement = abs(current - track)
            print(f"\nMove: |{current}-{track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")


        if right:
            total+=abs(current-right[0])
            print("\nJump from last left to first right")
            current = right[0]
            print(current, end=" ")


        for track in right[1:]:
            movement = abs(current - track)
            print(f"\nMove: |{current}-{track}| = {movement}")
            total += movement
            current = track
            print(track, end=" ")

    print("\nTotal head movement =", total)
queue = [98,183,37,122,14,124,65,67]
head = 53
direction ='i'
clook(queue, head, direction)
```

# Page Replacing Algorithms

```python
def fifo(ref, frames):
    memory = []
    miss = 0
    hit=0
    for page in ref:
        if page not in memory:
            if len(memory) < frames:
                memory.append(page)
            else:
                memory.pop(0)
                memory.append(page)
            miss += 1
        else:
          hit +=1
        print(page, "->", memory)

    print('\nTotal Miss =',miss)
    print('\n Total Hit =',hit)



pages = [1, 2, 3, 4, 1, 2, 5, 1, 2, 3, 4, 5]
frames = 3

fifo(pages, frames)
```

```python
def lru(ref, frames):
    memory = []
    miss = 0
    hit = 0

    for page in ref:

        if page in memory:
            hit += 1
            memory.remove(page)
            memory.append(page)
        else:
            miss += 1

            if len(memory) < frames:
                memory.append(page)
            else:
                memory.pop(0)
                memory.append(page)

        print(page, "->", memory)

    print("Miss =", miss)
    print("Hit =", hit)
pages = [1, 2, 3, 4, 1, 2, 5, 1, 2, 3, 4, 5]
frames = 3
lru(pages,frames)
```

```python
def optimal(pages, frames):

    memory = []
    miss = 0
    hits = 0

    for i in range(len(pages)):
        page = pages[i]

        if page in memory:
            hits += 1
            fault = "hit"

        else:

            miss += 1
            fault = "miss"

            if len(memory) < frames:
                memory.append(page)

            else:
                future = pages[i+1:]

                index_to_replace = -1
                farthest = -1

                for j in range(len(memory)):

                    if memory[j] not in future:
                        index_to_replace = j
                        break

                    else:
                        pos = future.index(memory[j])

                        if pos > farthest:
                            farthest = pos
                            index_to_replace = j

                memory[index_to_replace] = page

        print(f"Page: {page} -> Frames: {memory} -> {fault}")

    total = len(pages)
    hit_ratio = hits / total
    miss_ratio = miss / total

    print("\nTotal Pages:", total)
    print("Total Hits:", hits)
    print("Total Misses:", miss)
    print("Hit Ratio:", round(hit_ratio, 2))
    print("Miss Ratio:", round(miss_ratio, 2))


pages = [1, 2, 3, 4, 1, 2, 5, 1, 2, 3, 4, 5]
frames = 3

optimal(pages, frames)
```

# CPU Time Scheduling Algorithms

```python
def fcfs(processes, bt):
    n = len(processes)
    wt = [0] * n
    tat = [0] * n

    tat[0] = bt[0]

    for i in range(1, n):
        wt[i] = wt[i - 1] + bt[i - 1]
        tat[i] = wt[i] + bt[i]

    print("Process  BT  WT  TAT")
    for i in range(n):
        print(processes[i], bt[i], wt[i], tat[i])

    avg_wt = sum(wt) / n
    avg_tat = sum(tat) / n

    print("Average Waiting Time:", avg_wt)
    print("Average Turnaround Time:", avg_tat)
pro=[1,2,3,4]
br=[12,3,22,25]
fcfs(pro,br)
```

```python
def sjf(processes, bt):
    n = len(processes)

    for i in range(n):
        for j in range(i + 1, n):
            if bt[i] > bt[j]:
                bt[i], bt[j] = bt[j], bt[i]
                processes[i], processes[j] = processes[j], processes[i]

    wt = [0] * n
    tat = [0] * n

    for i in range(1, n):
        wt[i] = wt[i-1] + bt[i-1]

    for i in range(n):
        tat[i] = wt[i] + bt[i]

    print("Process  BT  WT  TAT")
    for i in range(n):
        print(processes[i], "     ", bt[i], " ", wt[i], " ", tat[i])

    avg_wt = sum(wt) / n
    avg_tat = sum(tat) / n

    print("Average Waiting Time:", avg_wt)
    print("Average Turnaround Time:", avg_tat)
pro=[1,2,3,4]
br=[12,3,22,25]
sjf(pro,br)
```

```python
def priority_scheduling(processes, bt, priority):
    n = len(processes)
    wt = [0] * n
    tat = [0] * n


    for i in range(n):
        for j in range(i + 1, n):
            if priority[i] > priority[j]:

                priority[i], priority[j] = priority[j], priority[i]

                bt[i], bt[j] = bt[j], bt[i]

                processes[i], processes[j] = processes[j], processes[i]

    tat[0] = bt[0]

    for i in range(1, n):
        wt[i] = wt[i - 1] + bt[i - 1]
        tat[i] = wt[i] + bt[i]

    print("Process  BT  Priority  WT  TAT")
    for i in range(n):
        print(processes[i], bt[i], priority[i], wt[i], tat[i])

    print("Average Waiting Time:", sum(wt)/n)
    print("Average Turnaround Time:", sum(tat)/n)


p=[1,2,3,4]
b=[12,3,22,25]
pr=[2,1,4,3]

priority_scheduling(p, b, pr)
```

```python
def round_robin(p, bt, tq):
    n = len(p)
    rem = bt.copy()
    wt = [0]*n
    time = 0
    e=[]
    e2=[]

    while sum(rem) > 0:
        for i in range(n):
            if rem[i] > 0:
                if rem[i] <= tq:
                    time += rem[i]
                    wt[i] = time - bt[i]
                    rem[i] = 0
                    e2.append(p[i])
                    e.append(time)
                else:
                    time += tq
                    rem[i] -= tq
                    e2.append(p[i])
                    e.append(time)

    print("Process  BT  WT  TAT")
    for i in range(n):
        tat = wt[i] + bt[i]
        print(p[i], bt[i], wt[i], tat)
    for i in range(len(e)):
        print(f'process {e2[i]} TAT {e[i]}')


    print("Average WT =", sum(wt)/n)
    print("Average TAT =", sum(wt[i] + bt[i] for i in range(n))/n)



p=[1,2,3,4]
bt=[12,3,22,25]
tq = 10

round_robin(p, bt, tq)
```
