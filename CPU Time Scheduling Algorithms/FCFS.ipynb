# First Come First Serve

```python
def fcfs(processes, bt):
    n = len(processes)
    wt = [0] * n
    tat = [0] * n

    tat[0] = bt[0]

    for i in range(1, n):
        wt[i] = wt[i - 1] + bt[i - 1]
        tat[i] = wt[i] + bt[i]

    print("Process  BT  WT  TAT")
    for i in range(n):
        print(processes[i], bt[i], wt[i], tat[i])

    avg_wt = sum(wt) / n
    avg_tat = sum(tat) / n

    print("Average Waiting Time:", avg_wt)
    print("Average Turnaround Time:", avg_tat)
```

```python
p = list(map(int, input("Enter numbers a: ").split()))
b = list(map(int, input("Enter numbers b: ").split()))

fcfs(p,b)
```
