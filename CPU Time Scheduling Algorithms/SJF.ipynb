# Shortest Job First (SJF)

```python
def sjf(processes, bt):
    n = len(processes)

    for i in range(n):
        for j in range(i + 1, n):
            if bt[i] > bt[j]:
                bt[i], bt[j] = bt[j], bt[i]
                processes[i], processes[j] = processes[j], processes[i]

    wt = [0] * n
    tat = [0] * n

    for i in range(1, n):
        wt[i] = wt[i-1] + bt[i-1]

    for i in range(n):
        tat[i] = wt[i] + bt[i]

    print("Process  BT  WT  TAT")
    for i in range(n):
        print(processes[i], "     ", bt[i], " ", wt[i], " ", tat[i])

    avg_wt = sum(wt) / n
    avg_tat = sum(tat) / n

    print("Average Waiting Time:", avg_wt)
    print("Average Turnaround Time:", avg_tat)

```

```python
p = list(map(int, input("Enter numbers a: ").split()))
b = list(map(int, input("Enter numbers b: ").split()))
sjf(p,b)
```
