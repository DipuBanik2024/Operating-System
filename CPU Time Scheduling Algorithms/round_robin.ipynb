# Round Robin

```python
def round_robin(p, bt, tq):
    n = len(p)
    rem = bt.copy()
    wt = [0]*n
    time = 0
    e=[]
    e2=[]

    while sum(rem) > 0:
        for i in range(n):
            if rem[i] > 0:
                if rem[i] <= tq:
                    time += rem[i]
                    wt[i] = time - bt[i]
                    rem[i] = 0
                    e2.append(p[i])
                    e.append(time)
                else:
                    time += tq
                    rem[i] -= tq
                    e2.append(p[i])
                    e.append(time)

    print("Process  BT  WT  TAT")
    for i in range(n):
        tat = wt[i] + bt[i]
        print(p[i], bt[i], wt[i], tat)
    for i in range(len(e)):
        print('prosses ',e2[i]+'TAT ',e[i])


    print("Average WT =", sum(wt)/n)
    print("Average TAT =", sum(wt[i] + bt[i] for i in range(n))/n)



p = input("Enter process names: ").split()
bt = list(map(int, input("Enter burst times: ").split()))
tq = int(input("Enter time quantum: "))

round_robin(p, bt, tq)
```
